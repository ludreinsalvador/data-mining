### **2. Pivot Tables**

```python
# EXPANSION 2: Pivot Tables (Fixed)
print("\n" + "="*60)
print("EXPANSION 2: PIVOT TABLES")
print("="*60)

# Pivot table: Average GWA by Course and Gender
pivot_gwa = df.pivot_table(values='last_term_GWA', 
                           index='course', 
                           columns='gender', 
                           aggfunc='mean')
print("Average GWA by Course and Gender:")
print(pivot_gwa.round(2))

# Pivot table: Payment status by Course
pivot_payment = df.pivot_table(values='student_number', 
                              index='course', 
                              columns='payment_status', 
                              aggfunc='count', 
                              fill_value=0)
print("\nStudent Count by Course and Payment Status:")
print(pivot_payment)

# Percentage of unpaid students per course (FIXED)
# Method 1: Calculate percentages manually
pivot_payment_pct = pivot_payment.div(pivot_payment.sum(axis=1), axis=0) * 100
print("\nPercentage Distribution by Course:")
print(pivot_payment_pct.round(1))

# Alternative method using crosstab with normalize
print("\nAlternative Percentage Calculation using crosstab:")
cross_tab_pct = pd.crosstab(df['course'], df['payment_status'], normalize='index') * 100
print(cross_tab_pct.round(1))

# Additional insights
print("\nKey Insights from Pivot Tables:")
print("=" * 40)

# Course with highest percentage of unpaid students
unpaid_pct = pivot_payment_pct['With Balance']
highest_unpaid = unpaid_pct.idxmax()
highest_unpaid_pct = unpaid_pct.max()

print(f"• Course with highest unpaid percentage: {highest_unpaid} ({highest_unpaid_pct:.1f}%)")
```

### **4. Alternative Predictive Analysis**

```python
# EXPANSION 4: Predictive Analysis (Without sklearn)
print("\n" + "="*60)
print("EXPANSION 4: PREDICTIVE ANALYSIS (NO SKLEARN)")
print("="*60)

# Create binary target: 1 for "With Balance", 0 for "Paid"
df['payment_binary'] = (df['payment_status'] == 'With Balance').astype(int)

# Encode categorical variables for analysis
df['gender_encoded'] = (df['gender'] == 'Female').astype(int)
df['scholarship_encoded'] = df['has_scholarship'].astype(int)

# Analysis 1: Payment probability by different factors
print("PAYMENT PROBABILITY ANALYSIS")
print("=" * 40)

# By Gender
payment_by_gender = df.groupby('gender')['payment_binary'].agg(['count', 'sum', 'mean'])
payment_by_gender.columns = ['Total_Students', 'Unpaid_Students', 'Unpaid_Probability']
print("Payment Probability by Gender:")
print(payment_by_gender.round(3))

# By Scholarship Status
payment_by_scholarship = df.groupby('has_scholarship')['payment_binary'].agg(['count', 'sum', 'mean'])
payment_by_scholarship.columns = ['Total_Students', 'Unpaid_Students', 'Unpaid_Probability']
print("\nPayment Probability by Scholarship Status:")
print(payment_by_scholarship.round(3))

# By Course
payment_by_course = df.groupby('course')['payment_binary'].agg(['count', 'sum', 'mean'])
payment_by_course.columns = ['Total_Students', 'Unpaid_Students', 'Unpaid_Probability']
print("\nPayment Probability by Course:")
print(payment_by_course.round(3))

# Analysis 2: Risk Factors Analysis
print("\n\nRISK FACTORS ANALYSIS")
print("=" * 40)

# Students with high units but unpaid
high_units_unpaid = df[(df['number_of_units_enrolled'] >= 21) & (df['payment_binary'] == 1)]
print(f"Students with ≥21 units and unpaid: {len(high_units_unpaid)} ({len(high_units_unpaid)/len(df)*100:.1f}%)")

# Students with low GWA and unpaid
low_gwa_unpaid = df[(df['last_term_GWA'] < 3.0) & (df['payment_binary'] == 1)]
print(f"Students with GWA <3.0 and unpaid: {len(low_gwa_unpaid)} ({len(low_gwa_unpaid)/len(df)*100:.1f}%)")

# Combined risk factors
high_risk = df[(df['last_term_GWA'] < 3.0) & (df['has_scholarship'] == False) & (df['payment_binary'] == 1)]
print(f"High-risk students (Low GWA + No Scholarship + Unpaid): {len(high_risk)} ({len(high_risk)/len(df)*100:.1f}%)")

# Analysis 3: Simple Correlation Analysis
print("\n\nCORRELATION WITH PAYMENT STATUS")
print("=" * 40)

# Calculate correlations with payment status
correlations = df[['payment_binary', 'gender_encoded', 'scholarship_encoded', 
                  'number_of_units_enrolled', 'last_term_GWA']].corr()['payment_binary'].sort_values(key=abs, ascending=False)

print("Correlation with Payment Status (1=Unpaid, 0=Paid):")
for var, corr in correlations.items():
    if var != 'payment_binary':
        direction = "Higher unpaid probability" if corr > 0 else "Lower unpaid probability"
        strength = "Strong" if abs(corr) > 0.3 else "Moderate" if abs(corr) > 0.1 else "Weak"
        print(f"  {var}: {corr:.3f} ({strength} - {direction})")

# Analysis 4: Predictive Rules (Simple Decision Tree Logic)
print("\n\nSIMPLE PREDICTIVE RULES")
print("=" * 40)

# Rule 1: Scholarship holders
scholarship_rule = df[df['has_scholarship'] == True]['payment_binary'].mean()
print(f"Rule 1 - Scholarship holders unpaid probability: {scholarship_rule:.3f}")

# Rule 2: High GWA students
high_gwa_rule = df[df['last_term_GWA'] >= 3.5]['payment_binary'].mean()
print(f"Rule 2 - High GWA (≥3.5) students unpaid probability: {high_gwa_rule:.3f}")

# Rule 3: Low units students
low_units_rule = df[df['number_of_units_enrolled'] <= 18]['payment_binary'].mean()
print(f"Rule 3 - Low units (≤18) students unpaid probability: {low_units_rule:.3f}")

# Combined rule prediction accuracy
print("\n\nPREDICTION ACCURACY TEST")
print("=" * 40)

# Simple prediction: Students with scholarship OR high GWA are likely to be paid
df['predicted_paid'] = ((df['has_scholarship'] == True) | (df['last_term_GWA'] >= 3.5)).astype(int)
df['predicted_unpaid'] = 1 - df['predicted_paid']

# Calculate accuracy
correct_predictions = (df['predicted_unpaid'] == df['payment_binary']).sum()
total_predictions = len(df)
accuracy = correct_predictions / total_predictions

print(f"Simple Rule Accuracy: {accuracy:.3f} ({correct_predictions}/{total_predictions})")
print("Rule: 'Students without scholarship AND low GWA (<3.5) are likely unpaid'")

# Confusion matrix
true_paid = ((df['payment_binary'] == 0) & (df['predicted_unpaid'] == 0)).sum()
false_paid = ((df['payment_binary'] == 1) & (df['predicted_unpaid'] == 0)).sum()
true_unpaid = ((df['payment_binary'] == 1) & (df['predicted_unpaid'] == 1)).sum()
false_unpaid = ((df['payment_binary'] == 0) & (df['predicted_unpaid'] == 1)).sum()

print(f"\nConfusion Matrix:")
print(f"True Paid (Correct): {true_paid}")
print(f"False Paid (Missed unpaid): {false_paid}")  
print(f"True Unpaid (Correct): {true_unpaid}")
print(f"False Unpaid (False alarm): {false_unpaid}")

print(f"\nPrecision (of unpaid predictions): {true_unpaid/(true_unpaid + false_unpaid):.3f}")
print(f"Recall (of actual unpaid): {true_unpaid/(true_unpaid + false_paid):.3f}")
```
